```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

#Carga de los datos en un dataframe del paquete Pandas

USA_Housing = "housing_train.csv"

df=pd.read_csv(USA_Housing)
```

```python

#Descripci√≥n de los datos

print('This data gives different sales prices with respect to type of houses in USA:')

#Comprobamos si el dataframe empieza y acaba igual

print(df.head())
print(df.tail())

#Resumen
print(df.info())
```

```python
#Creamos un objeto con los nombres de las columnas

headings = df.columns
print(headings)

#Averiguamos el tipo de dato de cada columna
tipos = df.dtypes
print(tipos)
```

```python
# Analizamos cuantitativamente (tenemos una lista SOLO con las columnas num√©ricas y sus estad√≠sticos)

anlsCuant=df.describe(include=[np.number])
anlsCuant
```

```python
#Describimos cualitativamente (lista con las variables cualitativas, con datos como la moda y la frecuencia)
anlsCual=df.describe(include=[object]) 
anlsCual
```

```python
#Creamos la matriz de correlaci√≥n

mcor=df.corr('pearson',1)
mcor
```

```python
#Gr√°ficamente usando el paquete Seaborn:
sns.heatmap(mcor, annot = True)
plt.show()
```

```python
#Vamos a rellenar la diagonal con valores NaN para que no aparezcan las autocorrelaciones:
np.fill_diagonal(mcor.values,np.nan)

#Paso la matriz a una serie, para luego ordenarla:

mcor_s = mcor.unstack()
print(mcor_s)
print('---------------------------------------------')

#Ordenamos

ordenados = mcor_s.sort_values()
print(ordenados)
```

```python
#Seleccionamos los que tengan una fuerte correlaci√≥n (>75%):

corr_fuerte = ordenados[abs(ordenados)>0.75]
corr_fuerte
```

```python
#De entre todas las correlaciones altas, la que m√°s me preocupa es la columna de OverallQual con el precio de venta
#que es lo que queremos predecir. Si entrenamos el modelo con esa fuerte relaci√≥n entre ambas, podr√≠a ser que a la hora de
#clasificar una casa sin esa variable, no diera buenos resultados. Por eso la voy a eliminar.

del df['OverallQual']
```

```python
#Buscamos qu√© variables tienen valores faltantes, y adem√°s, 
#eliminaremos aquellas cuyos valores faltantes superen el 10% del total de datos

print(df.info())

#Me quedo con las columnas que, como m√≠nimo, tengan el 90% del total de datos no nulos
df1=df.dropna(axis='columns',thresh=1314) 

#Vamos a ver c√≥mo han quedado el resto de variables y cu√°les tienen datos faltantes

print(df1.info())
```

```python
#Vamos a reemplazar los valores NaN de las variables relacionadas con el garaje con la categor√≠a none.
#con el df1.info() anterior, vemos que estas variables son las de ID 56 a 63.


#df1[['GarageType','GarageYrBlt','GarageFinish','GarageCars','GarageArea','GarageQual','GarageCond']].fillna(value=36,inplace=True)
df1.iloc[:,54:61] = df1.iloc[:,54:61].fillna('None')
df1.info()
```

```python
#Nos fijamos en la variable MasVnrArea, que es el area de revestimiento de mamposter√≠a en pies cuadrados

print('DESCRIPCI√ìN: \n',df1['MasVnrArea'].describe(),"\n")
print("LA MODA ES: ", df1['MasVnrArea'].mode(), "\n")
print("VALORES NULOS: ", df1['MasVnrArea'].isna().sum(),"\n")
print("FRECUENCIA DE LOS VALORES: \n", df1.MasVnrArea.value_counts(),'\n')


```

```python
#Como vemos, la moda de la variable es el 0, con 861 veces en 1460 registros, m√°s del 50%. Por tanto,
#decido reemplazar los valores nulos con la moda, que es el 0:

df1.MasVnrArea = df1.MasVnrArea.fillna(df1.MasVnrArea.mode()[0])
print ('VALORES NULOS: \n',df1['MasVnrArea'].isna().sum())
```

```python
#Vamos a elegir dos variables categ√≥ricas m√°s: BsmtExposure (paredes del s√≥tano a nivel de calle o de jard√≠n)
#y Electrical (sistema el√©ctrico):

print('DESCRIPCI√ìN BsmtExposure: \n',df1['BsmtExposure'].describe(),'\n')
print('CATEGOR√çAS BsmtExposure: \n',df1['BsmtExposure'].value_counts(),'\n')

print('DESCRIPCI√ìN Electrical: \n',df1['Electrical'].describe(),'\n')
print('CATEGOR√çAS Electrical: \n',df1['Electrical'].value_counts(),'\n')

#Como ambas modas tienen la frecuencia muy alta, 953 de 1422 en BsmtExposure y 1334 de 1459 en Electrictal, voy a rellenar 
#los valores nulos con la moda:

df1.BsmtExposure = df1.BsmtExposure.fillna(df1.BsmtExposure.mode()[0])
print ('VALORES NULOS BsmtExposure: \n',df1['BsmtExposure'].isna().sum())

df1.Electrical = df1.Electrical.fillna(df1.Electrical.mode()[0])
print ('VALORES NULOS Electrical: \n',df1['Electrical'].isna().sum())
```

```python
#Rellenamos los valores nulos del resto de variables tal y como indica el enunciado, num√©ricas con '0' 
#y categ√≥ricas con la moda:

for col in df1:
    #averiguamos el dtype de cada columna:
    
    dt = df1[col].dtype 
    
    #Comprobamos si es un num√©rica o categ√≥rica:
    
    if dt == int or dt == float:
        df1[col]=df1[col].fillna(0)
    elif dt==object:
        df1[col]=df1[col].fillna(df1[col].mode()[0])
        
#Por √∫ltimo, verificamos que todos los nulos han sido tratados, y el conjunto de datos est√° listo para procesar:

df1.info()
```

```python
## PROBLEMA DE REGRESI√ìN
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn import tree

#cargamos los datos en dos variables, datos completos en x y variable a predecir en y
x = df1
y = df1.SalePrice

#separamos en conjunto de entrenamiento y test. Adem√°s, convertimos las categor√≠as de las variables categ√≥ricas en nuevas variables, en este caso, num√©ricas,
#ya que scikit no acepta variables categ√≥ricas para los √°rboles de regresi√≥n

x_train, x_test, y_train, y_test = train_test_split(pd.get_dummies(x),y,test_size=0.2, random_state=42)

#creamos el √°rbol
arbol = DecisionTreeRegressor(random_state=42)

#buscamos los mejores par√°metros para el √°rbol:.
arbol_params = {'max_depth':np.arange(1,50,2),'min_samples_leaf':np.arange(2,15)}

from sklearn.model_selection import GridSearchCV

#especificamos el n√∫mero de divisiones que realizar√° la comprobaci√≥n k-fold, en este caso, hemos elegido k=10
gs_arbol = GridSearchCV(arbol,arbol_params,cv=10)
gs_arbol.fit(x_train,y_train)
bp = gs_arbol.best_params_
print (bp)
```

```python
#entrenamos el √°rbol con los par√°metros:

mejorArbol = DecisionTreeRegressor(max_depth=9, min_samples_leaf=2,random_state=42)
modelo = mejorArbol.fit(x_train,y_train)

#graficamos

fig = plt.figure(figsize=(50,50))
_ = tree.plot_tree(modelo,filled=True,feature_names=x_train.columns)

#como podemos observar, es un modeo muy grande, vamos a fijarnos con una profundidad de 2:

fig = plt.figure(figsize=(50,50))
_ = tree.plot_tree(modelo,filled=True, max_depth=2, feature_names=x_train.columns)
```

```python
#vamos a predecir la variable target:
y_pred=modelo.predict(x_test)

#y comprobamos el error cometido con el estimador MSE (Mean Squared Error)
from sklearn.metrics import mean_squared_error

print('MSE:',mean_squared_error(y_pred,y_test))
```
