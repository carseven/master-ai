```python
import numpy as np
import matplotlib.image as im
import matplotlib.pyplot as plt
import cv2
```

```python
def logImg(img,alpha):
    c=255/(np.log(1+(np.exp(alpha)-1)*np.max(img)))
    s = (np.zeros([img.shape[0],img.shape[1],img.shape[2]],dtype=np.uint8))
    s[:,:,0]=c*(np.log(1+((np.exp(alpha)-1)*img[:,:,0])))
    s[:,:,1]=c*(np.log(1+((np.exp(alpha)-1)*img[:,:,1])))
    s[:,:,2]=c*(np.log(1+((np.exp(alpha)-1)*img[:,:,2])))
    return s
```

```python
img = im.imread('Clase4_1.jpg')

alpha = 2
new_img = logImg(img,alpha)

gray=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
new_gray=cv2.cvtColor(new_img,cv2.COLOR_RGB2GRAY)
plt.figure()
f, axarr = plt.subplots(2,2) 
axarr[1][0].imshow(img)
axarr[1][1].imshow(new_img)
axarr[0][0].hist(gray)
axarr[0][1].hist(new_gray)

plt.show()
```

```python
def expImg(img,alpha):
    c=255/(np.power((1+alpha),np.max(img))-1)
    s = (np.zeros([img.shape[0],img.shape[1],img.shape[2]],dtype=np.uint8))
    s[:,:,0]=c*(np.power((alpha+1),img[:,:,0])-1)
    s[:,:,1]=c*(np.power((alpha+1),img[:,:,1])-1)
    s[:,:,2]=c*(np.power((alpha+1),img[:,:,2])-1)
    return s
```

```python
img = im.imread('Clase4_2.jpg')

alpha = 5.0
new_img = expImg(img,alpha)

gray=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
new_gray=cv2.cvtColor(new_img,cv2.COLOR_RGB2GRAY)
plt.figure()
f, axarr = plt.subplots(2,2) 
axarr[1][0].imshow(img)
axarr[1][1].imshow(new_img)
axarr[0][0].hist(gray)
axarr[0][1].hist(new_gray)
plt.show()
```
