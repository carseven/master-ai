# Importación de librerías e imagen

```python
# Se importan las librerías necesarias

import numpy as np
import cv2 as cv

from skimage import data, io
from scipy import ndimage as ndi
```

```python
# Cargamos la imagen del hombre con la cámara

img=data.camera()
io.imshow(img)
io.show()
```

# Sobel

```python
# Detecta bordes horizontales
Sx=np.array([[1, 2, 1],
       [0, 0, 0],
       [-1, -2, -1]])

# Detecta bordes verticales
Sy=np.array([[1, 0, -1],
       [2, 0, -2],
       [1, 0, -1]])
```

```python
# Gradiente x
Gx=ndi.convolve(img, Sx, output=np.float64, mode='nearest')

Gx_abs=np.sqrt(Gx**2)

norm_x = (Gx_abs - np.min(Gx_abs)) / (np.max(Gx_abs) - np.min(Gx_abs))

io.imshow(norm_x)
io.show()
```

```python
# Gradiente y
Gy=ndi.convolve(img, Sy, output=np.float64, mode='nearest')

Gy_abs=np.sqrt(Gy**2)

norm_y = (Gy_abs - np.min(Gy_abs)) / (np.max(Gy_abs) - np.min(Gy_abs))

io.imshow(norm_y)
io.show()
```

```python
G=np.sqrt(Gx**2+Gy**2)

norm = (G - np.min(G)) / (np.max(G) - np.min(G))

io.imshow(norm)
io.show()
```
